```python
import numpy as np 
import matplotlib.pyplot as plt      
import math
```

```python
class procent():
    def __init__(self, r, n, PV):
        self.r = r
        self.n = n
        self.PV = PV

    def prosty(r, n, PV):
        FV = PV*(1 + r*n)
        return FV

    def skladany(r, n, PV):
        FV = PV*(1 + r)**n
        return FV

class kombinatoryka():
    def __init__(self, n, k):
        self.n = n
        self.k = k

    def C(n, k):
        C = math.factorial(n)/(math.factorial(k) * math.factorial( n - k ))
        return C
```

```python
time = range(1, 9)

r1 = 0.10
V1 = 60000

V2 = 15000
r2 = 0.33

plt.plot(time , [ procent.prosty(r1, time[i], V1) for i in range(len(time)) ])
plt.plot(time , [ procent.skladany(r2, time[i], V2) for i in range(len(time)) ])
```
