```python
import matplotlib.pyplot as plt
from numpy import *
from scipy.interpolate import interp1d

x=[0, 1.6, 2.4, 3.4, 4.2, 5.0, 5.8]
y=[9, 2, 1.4, 1, 3.4, 8, 7.5]

f = interp1d(x, y)
f2 = interp1d(x, y, kind='quadratic')
xnew = linspace(0, 5.8, num=100, endpoint=True)


fig = plt.figure(figsize = (10,8) , dpi=80)
ax = plt.subplot()
ax.grid(True)
ax.set_axisbelow(True)

plt.scatter(x,y,label = 'Acquired data')
plt.plot(xnew, f2(xnew), '--', color = 'green', label = 'Estimated data')


ax.set_xlim(0,6)
ax.set_ylim(0, 9)
plt.legend(loc = 'best', fontsize = 12)
plt.xlabel('Loading steps [kN]', fontsize = 15)
plt.ylabel('RMS spectrum amplitude', fontsize = 15)

ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)
```
