```julia
using Plots
using DifferentialEquations
```

```julia
f(u,p,t) = 1.01*u
u0 = 1/2
tspan = (0.0,1.0)
prob = ODEProblem(f,u0,tspan)
sol = solve(prob, Tsit5(), reltol=1e-8, abstol=1e-8)

plot(sol,linewidth=1,
     xaxis="Time (t)",yaxis="u(t)", legend=false)
```
